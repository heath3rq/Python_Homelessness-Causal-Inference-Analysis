```python
import pandas as pd
```

```python
%%capture --no-display

data = pd.read_excel(
    "https://www.huduser.gov/portal/sites/default/files/xls/2007-2022-HIC-Counts-by-CoC.xlsx", 
    sheet_name=None, header=1)
```

```python
concatData = data[list(data.keys())[0]].copy()
concatData["Year"] = list(data.keys())[0]
for eachYear in list(data.keys())[1:-1]:
    data[eachYear]["Year"] = eachYear
    concatData = pd.concat((concatData, data[eachYear]))
concatData
```

```python
pit = pd.read_excel(
    "https://www.huduser.gov/portal/sites/default/files/xls/2007-2022-PIT-Counts-by-CoC.xlsx", sheet_name=None
)

concatPit = pit[list(pit.keys())[0]].copy()
concatPit["Year"] = list(pit.keys())[0]
for eachYear in list(pit.keys())[1:-1]:
    pit[eachYear]["Year"] = eachYear
    concatPit = pd.concat((concatPit, pit[eachYear]))
concatPit
```

```python
concatData.loc[
    concatData["CoC Number"] == "TX-700", "Total Year-Round Beds (RRH)"
].iloc[:9].reset_index(drop=True) / concatPit.loc[
    concatPit["CoC Number"] == "TX-700",
    concatPit.columns[
        [True if "Overall Homeless," in each else False for each in concatPit.columns]
    ],
].sum().iloc[
    :9
].reset_index(
    drop=True
)
```
